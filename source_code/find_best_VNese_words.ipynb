```python
import pandas as pd

train_df_clone = pd.read_csv('cleaned_train_df.csv')
```

```python
clean_comment_values = train_df_clone['clean_comment'].values
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

sw_file = "vietnamese-stopwords.txt"

with open(sw_file, "r", encoding = 'utf-8') as file:
        lines = file.readlines()

stop_words = [line.strip() for line in lines]

def delete_stop_words(text):
    tokens = text.lower().split()
    tokens = [t for t in tokens if t not in stop_words]
    return " ".join(tokens)

clone_train_df_v2 = train_df_clone.copy()
clone_train_df_v2['clean_v2_comment'] = clone_train_df_v2['clean_comment'].apply(lambda x : delete_stop_words(x))
clone_train_df_v2
```

```python
import ast

clone_train_df_v2['aspect'] = clone_train_df_v2['aspect'].apply(ast.literal_eval)
clone_train_df_v2['sentiments'] = clone_train_df_v2['sentiments'].apply(ast.literal_eval)

clone_train_df_exploded = clone_train_df_v2.explode(['aspect','sentiments'])
clone_train_df_exploded
```

```python
# Nh√≥m theo Aspect#Category
grouped = clone_train_df_exploded.groupby('aspect')['clean_v2_comment'].apply(list).to_dict()

# T·∫°o corpus to√†n b·ªô
all_sentences = clone_train_df_v2['clean_v2_comment'].tolist()

# D√πng TF-IDF
vectorizer = TfidfVectorizer(ngram_range=(1,2))  # ch·ªçn n-gram t·∫°i ƒë√¢y
vectorizer.fit(all_sentences)

# Vocabulary chung
vocab = vectorizer.get_feature_names_out()

# L∆∞u k·∫øt qu·∫£ t·ª´/ng·ªØ ti√™u bi·ªÉu cho m·ªói Aspect
top_keywords = {}

for aspect, sentences in grouped.items():
    tfidf_matrix = vectorizer.transform(sentences)
    mean_tfidf = tfidf_matrix.mean(axis=0).A1  # t√≠nh trung b√¨nh TF-IDF
    top_indices = mean_tfidf.argsort()[::-1][:10]  # ch·ªçn top 10
    top_terms = [(vocab[i], mean_tfidf[i]) for i in top_indices]
    top_keywords[aspect] = top_terms

# In k·∫øt qu·∫£
for aspect, terms in top_keywords.items():
    print(f"\nüîπ {aspect}:")
    for term, score in terms:
        print(f"  {term} ({score:.4f})")
```
